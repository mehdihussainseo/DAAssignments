```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CarsCompany.csv')
```

#### 1. Check and Drop Duplicate Records

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

#### 2. Check and Drop missing or null values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

#### 3. What are the top 10 average MSRP (Price) per car brand (rounded figure)? with Horizontal Bar Plot

```python
df.groupby('Make')['MSRP'].mean().round(decimals=2).sort_values(ascending=False).head(10).plot(kind='barh',title='Top 10 Car Brands & Avg Price')
plt.xlabel('Price In Millions')
plt.show()
```

#### 4. How has the MSRP (price) changed over the years (rounded figure)? with Line Plot

```python
df.groupby('Year')['MSRP'].mean().round(decimals=2).plot(kind='line')
plt.ylabel('Avg Price')
plt.show()
```

#### 5. How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

Hint: Average of (city mpg + highway MPG)

```python
df.groupby('Year')[['city mpg','highway MPG']].mean().plot(kind='line')
plt.ylabel('Average MPG')
plt.show()
```

#### 6. What percentage of cars are Compact, Midsize, or Large? Show with Pie chart

```python
df['Vehicle Size'].value_counts().plot(kind='pie')
plt.show()
```

#### 7. Which vehicle styles are sold the most? show with vertical bar plot

```python
df['Vehicle Style'].value_counts().plot(kind='bar')
plt.ylabel('No. of Units Sold')
plt.show()
```

#### 8. Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
df[(df['MSRP']>=40000) & (df['Popularity']>=1000)]['Make'].value_counts()
```
